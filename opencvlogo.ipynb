```python
img = np.zeros((512, 512, 3), np.uint8)
cv.line(img,(0,0),(511,511),(255,0,0),5)
cv.rectangle(img,(384,0),(510,128),(0,255,0),4)
cv.circle(img,(447,63), 63, (0,0,255), -1)
cv.ellipse(img,(256,256),(100,50),0,0,180,(20, 100, 150),-1)
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(img,[pts],True,(0,255,255))

mypts = np.array([[30, 30], [30, 50], [50, 50], [50, 30]])
mypts = mypts.reshape((-1, 1, 2))
cv.polylines(img, [mypts], True, (255, 255, 255))
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2,cv.LINE_AA)
cv2_imshow(img)
```

```python
opencvlogo = np.zeros((512, 512, 3), np.uint8)
angles = [-90, -45, 90]
colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]
innerRadius = 20
centers = np.array([(310, 300), (200, 300), (256, 210)], np.uint32)
for i in range(3):
  cv.ellipse(opencvlogo, centers[i], (50, 50), angles[i], 30, 330, colors[i], -1)
  cv.circle(opencvlogo, centers[i], innerRadius, (0, 0, 0), -1)
#cv.ellipse(opencvlogo, centers[i], (50, 50), -90, 30, 330, (255, 255, 255), -1)
cv2_imshow(opencvlogo)
```
